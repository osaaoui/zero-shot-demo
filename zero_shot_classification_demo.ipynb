## 1- Zero-shot-classification with "MoritzLaurer/DeBERTa-v3-base-mnli-fever-anli" model.
The data come from the consumer complaints of the CFPB's online database.


```python
zero_shot_2 = pipeline(
    task= "zero-shot-classification",
    model="MoritzLaurer/DeBERTa-v3-base-mnli-fever-anli"
)
```

```python
import pandas as pd
```

```python
def categorize_article(article):
  results= zero_shot_2(
      article,
      candidate_labels=["Credit reporting, credit repair services, or other personal consumer reports","Credit reporting or other personal consumer reports", "Money transfer, virtual currency, or money service", "Vehicle loan or lease", "Student loan","Credit cards", "Mortgage", "Checking and Savings account", "Debt collection"]
  )
  del results["sequence"]
  display(pd.DataFrame(results))
```
